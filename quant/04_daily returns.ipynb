```python
import pandas as pd
```

```python
data=pd.read_csv('00690.csv')
data.set_index(pd.to_datetime(data['Date']),inplace=True)
data.drop('Date',axis=1,inplace=True)
data.head()
```

# Daily Returns
## $$r_t= \frac{P_t-P_{t-1}}{P_{t-1}}$$

```python
df=data.copy()
df['daily_returns']=df['Price'].pct_change(1)
df.head()
```

```python
df['weekly_returns']=df['Price'].pct_change(5)
df.head(10)
```
