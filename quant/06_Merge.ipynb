```python
import pandas as pd
```

```python
ETF1=pd.read_csv('00690.csv')
ETF2=pd.read_csv('00728.csv')
```

# Merge
## Merge(data1,data2,依據,合併方式)
### 依據: on / left_on / right_on / left_index / right_index
### 合併方式: left / right / inner / outer

```python
df=pd.merge(ETF1,ETF2,on='Date',how='inner')
df.head()
```

```python
df.rename(columns={'Price_x':'00690','Price_y':'00728'},inplace=True)
df.head()
```

```python
ETF1.rename(columns={'Date':'time'},inplace=True)
ETF1.head()
```

```python
ETF2.head()
```

```python
df_2=pd.merge(ETF1,ETF2,left_on='time',right_on='Date',how='inner')
df_2.head()
```

## 用Index合併

```python
ETF1=pd.read_csv('00690.csv')
ETF1.set_index(pd.to_datetime(ETF1['Date']),inplace=True)
ETF1.drop('Date',axis=1,inplace=True)
ETF1.head()
```

```python
ETF2=pd.read_csv('00728.csv')
ETF2.set_index(pd.to_datetime(ETF2['Date']),inplace=True)
ETF2.drop('Date',axis=1,inplace=True)
ETF2.head()
```

```python
df_3=pd.merge(ETF1,ETF2,left_index=True,right_index=True,how='left')
df_3.tail(100)
```

```python
df_3.rename(columns={'Price_x':'00690','Price_y':'00728'},inplace=True)
df_3.head()
```
