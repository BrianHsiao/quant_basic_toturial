```python
import pandas as pd
from scipy.stats import linregress
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('00690.csv')
data.head()
```

```python
data.plot()
```

```python
df=data.copy()
reg_up=linregress(x=df.index,y=df.Price)
reg_up
```

```python
reg_up.slope
```

```python
reg_up[0]
```

```python
up_line=reg_up[1]+reg_up[0]*df.index
up_line
```

```python
df.Price.plot()
plt.plot(up_line)
```

```python
df_temp=df[df['Price']>up_line]
reg=linregress(x=df_temp.index,y=df_temp.Price)
df['HT']=reg[1]+reg[0]*data.index
```

```python
df.head()
```

```python
df.Price.plot()
plt.plot(df.HT)
```

```python
while len(df_temp)>=5:
    reg=linregress(x=df_temp.index,y=df_temp.Price)
    up_line=reg[1]+reg[0]*data.index
    df_temp=df[df['Price']>up_line]

df['High_Trend']=reg[1]+reg[0]*data.index
```

```python
df.head()
```

```python
df['Price'].plot()
df['High_Trend'].plot()
```
